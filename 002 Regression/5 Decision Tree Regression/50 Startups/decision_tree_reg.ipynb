```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("50_Startups.csv")
```

```python
x = df.iloc[:, : -1].values
y = df.iloc[:, -1].values
```

```python
x
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
colEncoder = ColumnTransformer(transformers=[("encoder", OneHotEncoder(), [3])], remainder="passthrough")
```

```python
x = colEncoder.fit_transform(x)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=0)
```

```python
regressor.fit(x_train, y_train)
```

```python
plt.plot(range(1, len(y_train) + 1 ),y_train, color="orange")
plt.plot(range(1, len(y_train) + 1), regressor.predict(x_train), color="blue")
```
