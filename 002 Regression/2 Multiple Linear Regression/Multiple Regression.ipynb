```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv("50_Startups.csv")
data_frame
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
col = ColumnTransformer(transformers=[("encoder", OneHotEncoder(), [3])], remainder="passthrough")
```

```python
x = data_frame.iloc[:, :-1].values
x
```

```python
con = np.array(col.fit_transform(x))
```

```python
con
```

```python
x = con
x
```

```python
y = data_frame.iloc[:, -1].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train)
```

```python
pred = regressor.predict(x_test)
```

```python
## Plot
plt.plot(range(1,len(y_test) + 1), y_test, color="red")
plt.plot(range(1,len(y_test) + 1), pred, color="blue")
plt.xticks(range(1,len(y_test) + 1))
plt.grid(True)
plt.xlabel("Profit")
plt.ylabel("IDK")
```

```python
np.set_printoptions(precision=2)
```

```python
differences = y_test - pred
```

```python
differences
```

```python
np.concatenate((y_test.reshape(len(y_test), 1), pred.reshape(len(pred) , 1), differences.reshape(len(y_test), 1)), 1)
```

```python
## Plot
plt.bar(range(1,len(y_test) + 1), differences)
plt.plot(range(1,len(y_test) + 1), y_test, color="red")
plt.plot(range(1,len(y_test) + 1), pred, color="blue")
plt.xticks(range(1,len(y_test) + 1))
plt.grid(True)
plt.xlabel("Profit")
plt.ylabel("IDK")
```
